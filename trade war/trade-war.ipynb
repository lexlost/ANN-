```python
import numpy as np
from scipy.stats import pearsonr

# 确保MFN_TAFF和GDP的形状是(85, 12)
assert MFN_TAFF.shape == (85, 11)
assert GDP.shape == (85, 11)

# 初始化相关性矩阵
correlation_matrix = np.zeros((11, 11))

# 计算每一列之间的相关性
for i in range(11):
    for j in range(11):
        correlation, _ = pearsonr(MFN_TAFF.iloc[:, i], GDP.iloc[:, j])
        correlation_matrix[i, j] = correlation

# 1. 绘制热图
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation between MFN_TAFF and GDP')
plt.xlabel('GDP Columns')
plt.ylabel('MFN_TAFF Columns')
plt.show()

# 2. 选择相关性最高的几对列绘制散点图
# 为简单起见，这里我只选择一个例子
max_corr = np.unravel_index(
    np.argmax(correlation_matrix, axis=None), correlation_matrix.shape)
sns.scatterplot(x=MFN_TAFF.iloc[:, max_corr[0]], y=GDP.iloc[:, max_corr[1]])
plt.title(
    f'Scatter Plot of Highest Correlation (MFN_TAFF Column {max_corr[0]} vs GDP Column {max_corr[1]})')
plt.xlabel(f'MFN_TAFF Column {max_corr[0]}')
plt.ylabel(f'GDP Column {max_corr[1]}')
plt.show()

# 3. 绘制线图展示某些列的相关性变化
# 选择具有代表性的列
selected_columns = [0, 1, 2]  # 您可以根据需要选择不同的列
for i in selected_columns:
    plt.plot(correlation_matrix[i, :], label=f'MFN_TAFF Column {i}')
plt.title('Correlation Trends of Selected MFN_TAFF Columns with All GDP Columns')
plt.xlabel('GDP Columns')
plt.ylabel('Correlation')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# 确保 MFN_TAFF 和 GDP 的形状是 (85, 11)
assert MFN_TAFF.shape == (85, 11)
assert GDP.shape == (85, 11)

# 初始化相关性矩阵
correlation_matrix = np.zeros((11, 11))

# 计算每一列之间的相关性
for i in range(11):
    for j in range(11):
        correlation, _ = pearsonr(MFN_TAFF.iloc[:, i], GDP.iloc[:, j])
        correlation_matrix[i, j] = correlation

# 假设 MFN_TAFF 和 GDP 已经是您的 DataFrame 数据
# MFN_TAFF.shape == (85, 11)
# GDP.shape == (85, 11)

# 创建一个新的 DataFrame，将 MFN_TAFF 和 GDP 合并
# 假设每列代表一个不同的年份
combined_data = pd.concat([MFN_TAFF, GDP], axis=1)

# 为了绘图的清晰性，我们可能需要确保年份标签是正确的
# 假设 MFN_TAFF 和 GDP 的列是年份
years = [f"Year {i}" for i in range(1, 12)] * 2  # 假设我们有11个年份，每个数据集各11个年份

# 生成线图
plt.figure(figsize=(12, 8))
for i, column in enumerate(combined_data.columns):
    plt.plot(combined_data.index, combined_data[column], label=years[i])

plt.title('Line Plot of MFN_TAFF and GDP Over Years')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(loc='upper left')  # 将图例放到左上角以避免遮挡图表
plt.tight_layout()  # 自动调整子图参数, 使之填充整个图像区域
plt.show()
```

```python
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# 确保MFN_TAFF和GDP的形状是(85, 11)
assert MFN_TAFF.shape == (85, 11)
assert GDP.shape == (85, 11)

# 初始化相关性列表
yearly_correlation = []

# 计算每一年的相关性
for i in range(11):
    correlation, _ = pearsonr(MFN_TAFF.iloc[:, i], GDP.iloc[:, i])
    yearly_correlation.append(correlation)

# 假设年份从2011年到2021年
years = np.arange(2011, 2022)

# 绘制相关性图像
plt.figure(figsize=(12, 6))
plt.plot(years, yearly_correlation, marker='o')

plt.title('Yearly Correlation between MFN_TAFF and GDP')
plt.xlabel('Year')
plt.ylabel('Pearson Correlation Coefficient')
plt.grid(True)
plt.xticks(years)
plt.tight_layout()
plt.show()
```
